```python
import pandas as pd
import numpy as np

from sklearn.preprocessing import StandardScaler, MinMaxScaler, LabelEncoder
from sklearn.ensemble import IsolationForest

import ppscore as pps
```

```python
df = pd.read_csv("adult_with_headers (1).csv")
df.head()
```

```python
df.info()
df.describe()
df.isnull().sum()
```

```python
df = df.replace(" ?", np.nan)

df.isnull().sum()
```

# Encoding Techniques

```python
for col in df.select_dtypes(include="object"):
    df[col].fillna(df[col].mode()[0], inplace=True)
```

```python
for col in df.select_dtypes(include=np.number):
    df[col].fillna(df[col].median(), inplace=True)
```

```python
num_cols = df.select_dtypes(include=np.number).columns

scaler = StandardScaler()
df_standard = df.copy()

df_standard[num_cols] = scaler.fit_transform(df_standard[num_cols])

df_standard.head()
```

```python
scaler = MinMaxScaler()

df_minmax = df.copy()

df_minmax[num_cols] = scaler.fit_transform(df_minmax[num_cols])

df_minmax.head()
```

```python
cat_cols = df.select_dtypes(include="object").columns
cat_cols
```

```python
small_cat = [c for c in cat_cols if df[c].nunique() < 5]
small_cat
```

```python
df = pd.get_dummies(df, columns=small_cat)
```

```python
large_cat = [c for c in cat_cols if c not in small_cat]
large_cat
```

```python
le = LabelEncoder()

for col in large_cat:
    df[col] = le.fit_transform(df[col])
```

```python
df["capital_total"] = df["capital_gain"] + df["capital_loss"]
```

```python
df["age_group"] = pd.cut(
    df["age"],
    bins=[0,25,40,60,100],
    labels=[0,1,2,3]
)
```

```python
df["age_group"] = df["age_group"].astype(int)
```

# Feature Engineering

```python
df["capital_gain"] = np.log1p(df["capital_gain"])
df["capital_loss"] = np.log1p(df["capital_loss"])
```

```python
iso = IsolationForest(contamination=0.01)

outliers = iso.fit_predict(df[num_cols])

df["outlier"] = outliers

df = df[df["outlier"] == 1]

df.drop("outlier", axis=1, inplace=True)
```
