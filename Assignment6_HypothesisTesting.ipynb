```python
import numpy as np
from scipy.stats import t, norm
```

## Step 1: Given Data

```python
sample_mean = 3050
X_mean = 600
sigma_X = 25
n = 25
```

## Step 2: Calculate Theoretical Mean and Standard Deviation

```python
mu = 1000 + 5 * X_mean
sigma_W = 5 * sigma_X

mu, sigma_W
```

## Step 3: Compute Z-Test Statistic

```python
z = (sample_mean - mu) / (sigma_W / np.sqrt(n))
z
```

## Step 4: Critical Value and P-value (α = 0.05)

```python
alpha = 0.05
z_critical = norm.ppf(1 - alpha)
p_value = 1 - norm.cdf(z)

z_critical, p_value
```

## Step 5: Decision

```python
if z > z_critical:
    print("Reject Null Hypothesis")
    print("Conclusion: Costs are higher than expected.")
else:
    print("Fail to Reject Null Hypothesis")
    print("Conclusion: No strong evidence of higher costs.")
```

# Confidence Interval Estimation - Print Head Durability

## Step 1: Data

```python
data = np.array([
    1.13, 1.55, 1.43, 0.92, 1.25,
    1.36, 1.32, 0.85, 1.07, 1.48,
    1.20, 1.33, 1.18, 1.22, 1.29
])

n = len(data)
mean = np.mean(data)
std_sample = np.std(data, ddof=1)

mean, std_sample
```

## Part A: 99% Confidence Interval (t-distribution)

```python
confidence = 0.99
alpha = 1 - confidence
df = n - 1

t_critical = t.ppf(1 - alpha/2, df)
margin_error_t = t_critical * (std_sample / np.sqrt(n))

ci_t = (mean - margin_error_t, mean + margin_error_t)
ci_t
```

## Part B: 99% Confidence Interval (z-distribution)

```python
sigma_population = 0.2

z_critical = norm.ppf(1 - alpha/2)
margin_error_z = z_critical * (sigma_population / np.sqrt(n))

ci_z = (mean - margin_error_z, mean + margin_error_z)
ci_z
```

## Final Conclusion

- Hypothesis testing suggests whether the cost model is valid.
- Confidence intervals estimate the true durability range.
- t-distribution is used when population standard deviation is unknown.
- z-distribution is used when population standard deviation is known.
