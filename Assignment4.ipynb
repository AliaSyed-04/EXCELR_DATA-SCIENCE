# Problem Statement : Hospital Patient Data Analysis

```python
import pandas as pd

df_patient = pd.read_csv("Patient_Data.csv")
df_billing = pd.read_csv("Billing_Data.csv")
df_patient.info()
```

```python
new_df_patient = df_patient.drop(labels=['ReceptionistID','CheckInTime'],axis=1)
new_df_patient
```

```python
total_bill_amount_per_department = new_df_patient.groupby('Department')['BillAmount'].sum()
total_bill_amount_per_department
```

```python
cleaned_patient_data = df_patient.drop_duplicates(subset='PatientID')
cleaned_patient_data
```

```python
cleaned_patient_data = cleaned_patient_data.copy()
cleaned_patient_data['BillAmount'] = cleaned_patient_data['BillAmount'].fillna(cleaned_patient_data['BillAmount'].mean())
cleaned_patient_data['BillAmount']
```

```python
df_merged = pd.merge(cleaned_patient_data, df_billing, on='PatientID')
df_merged
```

```python
df_new_patients = pd.DataFrame({
    'PatientID': [106, 107],
    'Name': ['Fiona', 'George'],
    'Department': ['Cardiology', 'Neurology'],
    'Doctor': ['Dr. Smith', 'Dr. John'],
    'BillAmount': [6800, 7200],
    'ReceptionistID': [2, 3],
    'CheckInTime': ['2023-01-15 10:00', '2023-01-16 11:30']
})

df_updated = pd.concat([cleaned_patient_data, df_new_patients], axis=0, ignore_index=True)
df_updated
```

```python
df_new_cols = pd.DataFrame({
    'InsuranceCovered': [2000, 1500, 3000, 2500, 1800, 1200, 1345],
    'FinalAmount': [3000, 4425, 4500, 3700, 4125, 1890, 2300]
})
df_updated = df_updated.drop(columns=['InsuranceCovered', 'FinalAmount'], errors='ignore')

df_updated_new_columns = pd.concat([df_updated, df_new_cols], axis=1)
df_updated_new_columns
```

# Expected Outcomes

Department wise Revenue

```python
df_updated_new_columns.groupby('Department')['FinalAmount'].sum()
```

Doctor's Performance

```python
df_updated_new_columns.groupby('Doctor')['FinalAmount'].sum()
```

Average Billing per Department

```python
df_updated_new_columns.groupby('Department')['BillAmount'].mean()
```
