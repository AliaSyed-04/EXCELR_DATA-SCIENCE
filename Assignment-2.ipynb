# Section A: Data Structures & Control Structures

```python
#creating a list of 5 integers.

list_1 = [3,2,1,4,5]

#append operation
append_operation = list_1.append(6)
print('After Append')
print(list_1)

#extend operation
list_extnd = [7,10]
extend_operation = list_1.extend(list_extnd)
print('After Extend')
print(list_1)

#insert operation
insert_operation = list_1.insert(4,6)
print('After Insert')
print(list_1)

#remove operation
remove_operation = list_1.remove(10)
print('After Remove')
print(list_1)

#pop operation
pop_operation = list_1.pop()
print('After Pop')
print(list_1)

#index operation
index_operation = list_1.index(1)
print('After Index')
print(index_operation)

#count operation
count_operation = list_1.count(6)
print('After Count')
print(count_operation)

#sort operation
sort_operation = list_1.sort()
print('After Sorting')
print(list_1)

#Reverse operation
reverse_operation = list_1.reverse()
print('After Reverse')
print(list_1)

#Clear Operation
clear_operation = list_1.clear()
print('After Clear')
print(list_1)
```

```python
#creating a tuple
tuple_1 = ('Sugar', 'Bugs Bunny', 'Sweet Pea')

#Answer
print('\n')
print("Answer: In a tuple, we cannot change the elements because it is immutable which means once the tuple is created, you cannot alter it's elements. But if we really want to change the elements then there is an indirect method. First we can convert the tuple to list, change the element using list operation and change it back into a tuple.")
print('\n')
print("In conclusion, we do not have a direct way to change the element of tuple.")
print('\n')
print('Below is the method how we can convert the tuple to a list, change the element and convert it back to a tuple.')

#converting this to a list
new_list = list(tuple_1)
new_list[1] = "Chickpea"

#convert this new list, back into a tuple
new_tuple = tuple(new_list)
print('\n')
print('original Tuple:', tuple_1)
print('new Tuple:',new_tuple)
```

```python
#creating a set using set function
sample_set = set([1,2,2,2,3,3,4,5,6,6,6])
print(sample_set)
print('\n')
print("A set removes the duplicate values and returns only the unique elements.")
```

```python
number = int(input('Enter your number: '))

if number > 0:
    if number%2 == 0:
        print(number, "is positive and even.")
    elif number%2 !=0:
        print(number, "is positive but not even. It's an Odd number.")
    elif number == 0:
        print('Number is Zero')
    elif number < 0:
        print('Number is Negative')
    else:
        print('The data entered is not right')
```
