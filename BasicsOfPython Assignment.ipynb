# Exercise 1: Prime Numbers
Write a Python program that checks whether a given number is prime or not. A prime number is a natural number greater than 1 
that has no positive divisors other than 1 and itself.


```python
num=int(input("Enter a number:"))
if num>1:
    prime = True
    for n in range(2,num):
        if num%n == 0:
            prime = False
            break
    if prime:
        print("Its a Prime Number")
    else:
        print("Not a Prime Number")
else:
    print("Not a Prime Number")
        
```

# Exercise 2: Product of Random Numbers
Develop a Python program that generates two random numbers and asks the user to enter the product of these numbers. The program should then check if the user's answer is correct and display an appropriate message.


```python
import random

num1 = random.randint(1,100)
num2 = random.randint(1,100)

print(num1)
print(num2)

product = int(input("Enter the product of these two numbers: "))

#Check if the product entered by the user is right or not
product_of_nums = num1*num2
if product_of_nums == product:
    print("You are Right. The answer is: ",product_of_nums)
else:
    print("You are Wrong. The answer is not:", product,". The Right answer is: ",product_of_nums)
```

# Exercise 4:
✅ Problem Statement (Question): Hospital Billing System
A hospital charges its patients based on the following:
•	Room charges per day: ₹2,000
•	Doctor consultation fee (flat): ₹1,500
•	Lab test charges: Based on the number of tests taken. Each test costs ₹300
•	Medicine charges: Total cost of medicines provided
•	Discount: If the total bill (before discount) exceeds ₹10,000, a 10% discount is applied
Write a Python program to calculate the final bill for a patient who:
•	Stayed for 4 days
•	Had 3 lab tests
•	Had medicine charges of ₹2,400


```python
room_charge = 2000
consultation_fee = 1500
lab_test = 300
discount = 0.1 #because 10% can be written as 10/100 which is equal to 0.1

#patient stayed in hospital for 4 days
in_patient_days = 4
#num of lab tests patient had is 3
num_of_lab_tests = 3
#Medicine Charge is 2400
meds_cost = 2400

cost_for_stay = in_patient_days*room_charge
total_lab_test_cost = num_of_lab_tests*lab_test

total_bill = meds_cost+cost_for_stay+total_lab_test_cost+consultation_fee

if total_bill > 10000:
    applied_discount = discount*total_bill
    bill_after_discount = total_bill - applied_discount
    print("Your bill exceeds 10000Rs. A Discount of 10% will be applied. Bill after applying Discount: ", bill_after_discount)
else:
    print("Your total bill is: ",total_bill)

```
